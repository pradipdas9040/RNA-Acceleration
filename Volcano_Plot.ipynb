```python
import pandas as pd
import numpy as np
from bioinfokit import analys, visuz
```

```python
df = pd.read_csv("/content/volcano.csv")
df2 = pd.read_csv("/content/expgene.csv")
df
```

```python
df2
```

```python
visuz.GeneExpression.volcano(df=df, lfc='l2fc', pv='pv', geneid="Gene", genenames=('Setbp1','Plcb1','St18','Prdm16','Fkbp9','Alad', 'Btg2'), gstyle=2)
```

```python
lfc_thr = (1, 1)
pv_thr=(0.05, 0.05)
color=("green", "grey", "red")
lfc = 'l2fc'
pv = 'pv'
```

```python
df.loc[(df[lfc] >= lfc_thr[0]) & (df[pv] < pv_thr[0]), 'color_add_axy'] = color[0]  # upregulated
df.loc[(df[lfc] <= -lfc_thr[1]) & (df[pv] < pv_thr[1]), 'color_add_axy'] = color[2]  # downregulated
df['color_add_axy'].fillna(color[1], inplace=True)  # intermediate
df['logpv_add_axy'] = -(np.log10(df[pv]))
```

```python
df
```

```python
df3 = df[df.color_add_axy=='green']
df3.sort_values(by=['pv'], inplace=True)
up_reg = df3.head(250)
up_reg.to_csv("upregulated.csv")
```

```python
df4= df[df.color_add_axy=='red']
df4.sort_values(by=['pv'], inplace=True)
down_reg = df4.head(250)
down_reg.to_csv("downregulated.csv")
```
